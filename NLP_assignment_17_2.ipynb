```python
from transformers import pipeline
import pandas as pd
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

class CustomTranslator:
    def __init__(self, model_name):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

    def __call__(self, text):
        inputs = self.tokenizer(text, return_tensors="pt", padding=True)
        # A default max_length is used, it can be adjusted if needed.
        outputs = self.model.generate(**inputs, max_length=512)
        translated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        return [{"translation_text": translated_text}]

translator_fr = CustomTranslator("Helsinki-NLP/opus-mt-en-fr")
translator_de = CustomTranslator("Helsinki-NLP/opus-mt-en-de")
translator_hi = CustomTranslator("Helsinki-NLP/opus-mt-en-hi")

print("Models Loaded Successfully")
```

```python
df = pd.read_csv("/content/eng_-french.csv")
df
```

```python
sentences = df["English words/sentences"].tolist()

print(sentences[:5])
```

```python
french_translations = []

for sentence in sentences:
    translated = translator_fr(sentence)[0]['translation_text']
    french_translations.append(translated)

df["French"] = french_translations

df.head()
```

```python
german_translations = []
hindi_translations = []

for sentence in sentences:
    german = translator_de(sentence)[0]['translation_text']
    hindi = translator_hi(sentence)[0]['translation_text']

    german_translations.append(german)
    hindi_translations.append(hindi)

df["German"] = german_translations
df["Hindi"] = hindi_translations

df
```

```python
df.to_csv("translated_output.csv", index=False)

print("File Saved Successfully")
```
