```python
"""
This cell imports all required libraries.
spaCy is used for NLP processing.
pandas is used for tabular representation.
"""

import spacy
import pandas as pd
from spacy import displacy
```

```python
"""
This cell loads the English language model.
Language models are required because they contain:
- Vocabulary
- Grammar rules
- Statistical patterns for parsing
"""

# Load small English model
nlp = spacy.load("en_core_web_sm")

print("Model loaded successfully!")
```

```python
"""
This cell defines the dataset.
We use 15 sentences of different types:
- Simple
- Compound
- Complex
- Interrogative
- Passive
"""

sentences = [
    "The boy plays cricket.",
    "She is reading a book.",
    "The dog barked loudly at night.",
    "I went to the market and bought fruits.",
    "He studied hard because he wanted to pass.",
    "What are you doing?",
    "The cake was eaten by the child.",
    "Artificial intelligence is transforming industries.",
    "The teacher explained the concept clearly.",
    "They are watching a movie.",
    "The car was repaired by the mechanic.",
    "If it rains, we will cancel the trip.",
    "She sings and dances beautifully.",
    "The quick brown fox jumps over the lazy dog.",
    "Engineers design systems that solve problems."
]

print("Total sentences:", len(sentences))
```

```python
"""
This cell processes sentences using spaCy pipeline.

spaCy pipeline performs:
1. Tokenization → Splits sentence into words
2. POS Tagging → Assigns grammatical tags
3. Dependency Parsing → Identifies relationships
"""

docs = [nlp(sentence) for sentence in sentences]

print("All sentences processed successfully!")
```

```python
"""
This function extracts syntactic details from a sentence.

Parameters:
doc (spaCy Doc object)

Returns:
DataFrame containing:
- Token
- Lemma
- POS Tag
- Dependency
- Head word
"""

def extract_dependency_info(doc):
    data = []

    for token in doc:
        # Append token-level linguistic features
        data.append([
            token.text,          # actual word
            token.lemma_,        # base form
            token.pos_,          # part of speech
            token.dep_,          # dependency relation
            token.head.text      # head word
        ])

    # Convert into DataFrame
    df = pd.DataFrame(data, columns=["Token", "Lemma", "POS", "Dependency", "Head"])
    return df
```

```python
"""
This cell displays dependency tables for each sentence.
"""

for i, doc in enumerate(docs):
    print(f"\nSentence {i+1}: {doc.text}")

    df = extract_dependency_info(doc)
    display(df)
```

```python
"""
This function identifies:
- Subject
- Verb (ROOT)
- Object
- Modifiers
"""

def analyze_sentence(doc):
    subject = []
    obj = []
    root = None
    modifiers = []

    for token in doc:
        # Identify subject
        if "subj" in token.dep_:
            subject.append(token.text)

        # Identify object
        if "obj" in token.dep_:
            obj.append(token.text)

        # Identify root verb
        if token.dep_ == "ROOT":
            root = token.text

        # Identify modifiers
        if token.dep_ in ["amod", "advmod"]:
            modifiers.append(token.text)

    return {
        "Subject": subject,
        "Verb": root,
        "Object": obj,
        "Modifiers": modifiers
    }


# Apply analysis
for doc in docs:
    print("\nSentence:", doc.text)
    print(analyze_sentence(doc))
```

```python
"""
This cell visualizes dependency trees using displaCy.

We visualize first 5 sentences.
"""

for doc in docs[:5]:
    displacy.render(doc, style="dep", jupyter=True)
```
