```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import pandas as pd
```

```python
model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def summarize_text(text, max_length=150, min_length=30, do_sample=False):
    inputs = tokenizer(text, max_length=1024, truncation=True, return_tensors="pt")
    summary_ids = model.generate(
        inputs["input_ids"],
        num_beams=4,
        max_length=max_length,
        min_length=min_length,
        do_sample=do_sample,
        early_stopping=True,
        forced_bos_token_id=tokenizer.bos_token_id
    )
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
data = pd.read_csv("/content/summarization_dataset_lab18.csv")

print(data.head())
```

```python
texts = data["Text"].tolist()
```

```python
for text in texts:
    summary = summarize_text(text, max_length=120, min_length=40, do_sample=False)

    print("ORIGINAL TEXT:\n", text)
    print("\nSUMMARY:\n", summary)
    print("\n---------------------------------------------------\n")
```

```python
summary_short = summarize_text(texts[0], max_length=50, min_length=20, do_sample=False)
print(summary_short)
```

```python
summary_medium = summarize_text(texts[0], max_length=100, min_length=40, do_sample=False)
print(summary_medium)
```

```python
summary_long = summarize_text(texts[0], max_length=150, min_length=70, do_sample=False)
print(summary_long)
```

```python
long_text = texts[0] + texts[1]

summary = summarize_text(long_text, max_length=150, min_length=60, do_sample=False)

print(summary)
```
